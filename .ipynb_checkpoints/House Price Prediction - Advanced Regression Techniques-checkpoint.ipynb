```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

**1.Drop null values**

```python
data.head()
```

```python
newd = data.drop(['LotFrontage', 'Alley', 'MasVnrType', 'MasVnrArea', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Electrical', 'FireplaceQu', 'GarageType', 'GarageYrBlt',
       'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC', 'Fence',
       'MiscFeature'],axis = 1)
```

```python
len(newd.columns)
```

**2.Seperate qualitative and quantitative attributes**

quantitative part

```python
numd = newd.select_dtypes(np.number)
numd.head()
```

qualitative part

```python
catd = newd.select_dtypes(object)
catd.head()
```

**3.Feature selection on quantitative attributes**

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
X = numd.drop('SalePrice',axis = 1)
y = numd['SalePrice']
```

```python
bestfeat = SelectKBest(score_func = chi2,k = 10)
fit = bestfeat.fit(X,y)
scores = pd.DataFrame(fit.scores_)
cols = pd.DataFrame(X.columns)
featscores = pd.concat([cols,scores],axis = 1)
```

```python
featscores.columns = ['specs','scores']
print(featscores.nlargest(10,'scores'))
```

```python
featscores.nlargest(10,'scores')['specs'].to_list()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
ext =  ExtraTreesRegressor()
ext.fit(X,y)
```

```python
print(ext.feature_importances_)
```

```python
feat_importances = pd.Series(ext.feature_importances_,index = X.columns)
```

```python
feat_importances.nlargest(10).plot(kind = 'barh')
```

```python
newd = newd[['OverallQual','GarageCars','GrLivArea']]
```

**The features we can select are OverallQual,GarageCars,GRLIvArea,FullBath**

**4.Feature selection on qualitative attributes**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OrdinalEncoder
from sklearn.feature_selection import SelectFromModel
rfc = RandomForestClassifier()
oe = OrdinalEncoder()
```

```python
def prepare(X):
    oe.fit(X)
    X_enc = oe.transform(X)
    X_enc = pd.DataFrame(X_enc)
    X_enc.columns = X.columns
    return X_enc
catd_enc = prepare(catd)
```

```python
catd.head()
```

```python
catd_enc.head()
```

```python
rfc.fit(catd_enc,y)
```

```python
feat_importances = pd.Series(rfc.feature_importances_,index = catd_enc.columns)
```

```python
feat_importances.nlargest(10).plot(kind = 'barh')
```

**We can take Neighborhood,HouseStlpe,Exterior2nd,Exterior1st**

**5.Corelation matrix for quantitative attributes and drop high correlated attributes**

```python
catd = catd[['Neighborhood','HouseStyle','Exterior2nd','Exterior1st']]
```

```python
fulld = pd.concat([newd,catd],axis = 1)
fulld = pd.concat([fulld,y],axis = 1)
fulld.head()
```

```python
sns.scatterplot(x = 'GrLivArea',y = 'SalePrice',hue = 'HouseStyle',data = fulld)
```

```python
fulld_corr = fulld.corr()
sns.heatmap(fulld_corr)
```

**6.Encode qualitative attributes**

```python
fulld1 = pd.get_dummies(fulld[['Neighborhood','HouseStyle','Exterior2nd','Exterior1st']],drop_first = True)
```

```python
fulld1 = pd.DataFrame(fulld1)
```

```python
fulld1.head()
```

```python
fulld.drop(['Neighborhood','HouseStyle','Exterior2nd','Exterior1st'],axis = 1,inplace = True)
```

**7.Create new data frame by concatenating selected Qualitative and Quantitative attributes**

```python
fulld = pd.concat([fulld,fulld1],axis = 1)
fulld.head()
```

**8.Apply Machine learning algorithm to predict**

```python
from sklearn.model_selection import train_test_split
X = fulld.drop('SalePrice',axis = 1)
y = fulld['SalePrice']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.svm import SVR
svr = SVR(epsilon = 0.01,C = 1000,kernel = 'linear',gamma = 'auto')
svr.fit(X_train,y_train)
```

```python
svr.score(X_test,y_test)
```

```python
sns.distplot(y_test - svr.predict(X_test))
```
